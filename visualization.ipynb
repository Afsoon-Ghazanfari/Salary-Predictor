```python
df['YearsCodePro'].isnull().sum()
```

```python
plt.figure(figsize=(8,9))
sort_order = [float('nan'), 'Less than 1 year'] + [str(i) for i in range(50)] + ['More than 50 years']
df["sort_order"] = df.YearsCodePro.apply(lambda i:-1 if i not in sort_order else sort_order.index(i))
df = df.sort_values(by="sort_order")
sns.barplot(df,y='YearsCodePro',x='CompTotal')
plt.title('Years of experience vs Salary')
```

#Plot multiple selection questions vs Salary

```python
employment_type = []
for ix, row in df.query("Employment.notna() and CompTotal.notna()").iterrows():
    e = row["Employment"]
    c = row["CompTotal"]

    for _e in e.split(";"):
        new_row = [_e, c]
        employment_type.append(new_row)

employment_type = pd.DataFrame(employment_type, columns=["Employment", "Salary"])
sns.barplot(employment_type, y="Employment", x="Salary")
plt.title('Employment vs Salary')
```

```python
program_lenguage = []
for ix, row in df.query("LanguageHaveWorkedWith.notna() and CompTotal.notna()").iterrows():
    l = row["LanguageHaveWorkedWith"]
    c = row["CompTotal"]

    for _l in l.split(";"):
        new_row = [_l, c]
        program_lenguage.append(new_row)

program_lenguage = pd.DataFrame(program_lenguage, columns=["programLanguage", "Salary"])
plt.figure(figsize=(6,10))
sns.barplot(program_lenguage, y="programLanguage", x="Salary")
plt.title('programming Language vs Salary')
```

```python
frame_work = []
for ix, row in df.query("WebframeHaveWorkedWith.notna() and CompTotal.notna()").iterrows():
    f = row["WebframeHaveWorkedWith"]
    c = row["CompTotal"]

    for _f in f.split(";"):
        new_row = [_f, c]
        frame_work.append(new_row)

frame_work = pd.DataFrame(frame_work, columns=["FrameWork", "Salary"])
plt.figure(figsize=(6,10))
sns.barplot(frame_work, y="FrameWork", x="Salary")
plt.title('Frame Work vs Salary')
```
