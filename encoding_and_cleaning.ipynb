```python
#Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
#Creating Data Frame
df_raw=pd.read_csv('/content/drive/My Drive/survey_results_germany_modified.csv')
df=df_raw[df_raw['CompTotal'].notnull()]
df=df[df['Country']=='Germany']
df.head()
```

```python
df.columns
```

```python
#creat new data frame with some columns
df = df[['Employment','RemoteWork','EdLevel','YearsCodePro','OrgSize','CompTotal']]
df.head()
```

```python
#df = df.drop_duplicates(keep='first')
```

```python
#number of null values
df.isnull().sum()
```

```python
df.shape
```

```python
#count of each value
df["Employment"].value_counts()
```

```python
#filtering with employment full time values
df=df[df["Employment"]=='Employed, full-time']
df = df.drop("Employment",axis=1)
df.info()
```

```python
#change strings to integers
df['YearsCodePro']=df['YearsCodePro'].replace('Less than 1 year',0).astype(int)
df['YearsCodePro']=df['YearsCodePro'].replace('More than 50 years',51).astype(int)
```

```python
df=df.reset_index()
```

```python
df=df.drop(columns=['index'])
```

```python
#onehotencoding organisation size
enc = OneHotEncoder(sparse=False)
columns_to_one_hot = ['OrgSize']
encoded_array = enc.fit_transform(df.loc[:,columns_to_one_hot])
df_encoded = pd.DataFrame(encoded_array,columns=enc.get_feature_names_out() )
df_sklearn_encoded = pd.concat([df,df_encoded],axis=1)
df_sklearn_encoded.drop(labels= columns_to_one_hot,axis=1,inplace=True)
df_sklearn_encoded.head()
```

```python
df.info()
```

```python
#onehotencoding education level
enc = OneHotEncoder(sparse=False)
columns_to_one_hot = ['EdLevel']
encoded_array = enc.fit_transform(df_sklearn_encoded.loc[:,columns_to_one_hot])
df_encoded = pd.DataFrame(encoded_array,columns=enc.get_feature_names_out() )
df_sklearn_encoded = pd.concat([df_sklearn_encoded,df_encoded],axis=1)
df_sklearn_encoded.drop(labels= columns_to_one_hot,axis=1,inplace=True)
df_sklearn_encoded.head()
```

```python
#onehotencoding remote work
enc = OneHotEncoder(sparse=False)
columns_to_one_hot = ['RemoteWork']
encoded_array = enc.fit_transform(df_sklearn_encoded.loc[:,columns_to_one_hot])
df_encoded = pd.DataFrame(encoded_array,columns=enc.get_feature_names_out() )
df_sklearn_encoded = pd.concat([df_sklearn_encoded,df_encoded],axis=1)
df_sklearn_encoded.drop(labels= columns_to_one_hot,axis=1,inplace=True)
df_sklearn_encoded.head()
```
